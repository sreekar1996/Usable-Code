```python
def avg1(dataframe):
    if(dataframe["variable"]=="srk"):
        return((dataframe["mark2"]/dataframe["marks"])*dataframe["inflation"])
    else:
        return(dataframe["marks"]*dataframe["inflation"])
```

```python
def avg(dataframe,i):
    if(dataframe["variable"]=="srk"):
        return((dataframe["mark2"]/dataframe["marks"])*i)
    else:
        return(dataframe["marks"]*i)
```
